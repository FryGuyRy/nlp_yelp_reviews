```python
scategories=[cat for cat in cats]
keywords = ['Restaurants','Food','Bars']
ind = set(ind for ind, cat in enumerate(scategories) for word in cat if word in keywords)
len(ind)
```

```python
restaurant_ids = data_business['business_id'].iloc[list(ind)]
print(len(restaurant_ids))
restaurant_ids.head()
```

```python
restaurant_df = data_reviews[data_reviews['business_id'].isin(restaurant_ids)]
print(restaurant_df.describe())
print(restaurant_df.info())
restaurant_df.head()
```

```python
text, review_rating = [], []
for i in range(restaurant_df.shape[0]):
    text.append(restaurant_df['text'].iloc[i])
    review_rating.append(restaurant_df['stars'].iloc[i])
print(len(text), len(review_rating))
text[1], review_rating[1]
```
